```python
import numpy as np

corpus = []
with open('teks.txt') as f:
    for corpuss in f:
        corpus.append(corpuss)

from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english')
response = vectorizer.fit_transform(corpus)
print(response)
```
